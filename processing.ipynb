```python
import pandas as pd
import numpy as np
```

```python
file_path = 'Data/raw/01-2018-30-2018-banking-statment-dataset.csv'
try:
    if file_path.endswith('.csv'):
        df = pd.read_csv(file_path)
    else:
        raise ValueError("Unsupported file format. Supported formats are: .csv, .xlsx, .xls, .json")
except FileNotFoundError:
    raise FileNotFoundError(f"The file at {file_path} was not found.")
except Exception as e:
    raise RuntimeError(f"An error occurred while reading the file: {e}")

df
```

```python
df.dropna()
```

```python
category_sums = df.groupby('Category', as_index=False)[['Debt Amount', 'Credit Amount']].sum()
category_sums
```

```python
total_credit = category_sums['Credit Amount'].sum()
total_debt = category_sums['Debt Amount'].sum()

# Calculate percentages
category_sums['Credit percentage'] = (category_sums['Credit Amount'] / total_credit) * 100
category_sums['Debt percentage'] = (category_sums['Debt Amount'] / total_debt) * 100
category_sums
```

```python
df.to_csv('Data/outputs/category_summary.csv', index=False)
```
